# Expolatory Data Analysis  

## Gender : 

```python
df.columns
```

```python
ax  = sns.countplot(x= "Gender",data = df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)["Amount"].sum().sort_values(by = "Amount",ascending = False)
sns.barplot(x = "Gender",y = "Amount" , data  =sales_gen )
```

From Above Graphs we can see that most of the buyers are females and even the perchasing power of females are greterthan men

## Age 

```python
df.columns
```

```python
ax  = sns.countplot(x = "Age Group" , hue  = "Gender",data = df )
```

```python
ax  = sns.countplot(x = "Age Group" , hue  = "Gender",data = df )
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(["Age Group"],as_index = False)["Amount"].sum().sort_values(by = "Amount",ascending = False)
sns.barplot(x = "Age Group" , y = "Amount" , data = sales_age )
```

From Above  the Graphs we can see most of the buyers are of Age Group Between 26-35 years female

## State : 

```python
# Total Numbers of Orders From top 10 states

sales_states = df.groupby(["State"],as_index = False)["Orders"].sum().sort_values(by = "Orders",ascending = False).head(10)
sns.set(rc = {"figure.figsize" : (15,6)})

sns.barplot(x = "State" , y = "Orders" , data = sales_states)
```

```python
# Total Numbers of sales\Amount From top 10 states

sales_states = df.groupby(["State"],as_index = False)["Amount"].sum().sort_values(by = "Amount",ascending = False).head(10)
sns.set(rc = {"figure.figsize" : (15,6)})
c = ["r","y","g","b","r","g","y","b","g","p"]
sns.barplot(x = "State" , y = "Amount" , data = sales_states)
```

## Merital Status

```python
ax = sns.countplot(x = 'Shadi',data = df)
sns.set(rc = {"figure.figsize" : (7,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_state = df.groupby(["Shadi","Gender"],as_index = False)["Amount"].sum().sort_values(by = "Amount",ascending = False)
sns.barplot(x = "Shadi" , y = "Amount", data = sales_state,  hue= "Gender" )
```

 From Above Graphs we can see that most of the buyers are married (Women) and they have high purchasing power

 ##  Occupation 

```python
sns.set(rc = {"figure.figsize" : (20,5)})
ax = sns.countplot(x = 'Occupation',data = df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_Occupation = df.groupby(["Occupation"],as_index = False)["Amount"].sum().sort_values(by = "Amount",ascending = False)
sns.barplot(x = "Occupation" , y = "Amount" ,  data = sales_Occupation)
```

From Above Graphs we can see that most of the buyers are working in  IT sector,Healthcare And Aviation . 

## Product Catagory 

```python
df.columns
```

```python
sns.set(rc = {"figure.figsize" : (25,5)})
ax = sns.countplot(x = 'Product_Category',data = df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sns.set(rc = {"figure.figsize" : (19,5)})
sales_Category = df.groupby(["Product_Category"],as_index = False)["Amount"].sum().sort_values(by = "Amount",ascending = False).head(10)
sns.barplot(x = "Product_Category" , y = "Amount" ,  data = sales_Category)
```

From Above Graphs we can see that most of the sold products are Food , Clothing& Apparel , Electonic& Gadgets

```python
sns.set(rc = {"figure.figsize" : (19,5)})
sales_Category = df.groupby(["Product_ID"],as_index = False)["Orders"].sum().sort_values(by = "Orders",ascending = False).head(10)
sns.barplot(x = "Product_ID" , y = "Orders" ,  data = sales_Category)
```
